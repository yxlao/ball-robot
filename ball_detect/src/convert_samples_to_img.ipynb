```python
import numpy as np
import cv2
from os import listdir
from os.path import isfile, join
```

```python
sample_path = 'samples'
files = [f for f in listdir(sample_path) 
         if isfile(join(sample_path, f)) and f[-3:] == 'npy']
```

```python
for f in files:
    print f
    im_hsv = np.load(join(sample_path, f))
    cv2.imwrite(join(sample_path, f[:-3] + 'png'), im_hsv)
```
